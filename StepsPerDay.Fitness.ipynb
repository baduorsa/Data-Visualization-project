```python
import pandas as pd
```

```python
df = pd.DataFrame({'StepsPerDay': [12000,11000,10000,9000,8000,7000,6000,5000,4000,3000],'FitnessScore': [95,91,88,82,78,70,65,60,50,45]})
```

```python
x=df['StepsPerDay'].values.reshape (-1,1)
y=df['FitnessScore']
```

```python
 from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
bb=LinearRegression()
```

```python
bb.fit(x_train,y_train)
```

```python
y_pred=bb.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print ('Mean Squared Error(MSE):' ,mse)
print('r2 Score' ,r2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test,y_test, color='magenta',label='Actual Values')
plt.plot(x_test, y_pred, color='blue',label='Predicted Values')
plt.xlabel('Steps Per Day')
plt.ylabel('Fitness Score')
plt.legend()
plt.show()
```
